> 터들로 카드짝 맞추기 게임 만들기


```python
#터들로 카드짝 맞추기 게임 만들기
import turtle as t
import random
import time
import winsound

def find_card(x,y):
    min_idx=0
    min_dis=100

    for i in range(16):
        distance=turtles[i].distance(x,y)
        if distance<min_dis:
            min_dis=distance
            min_idx=i
    return min_idx

def result(m):
    t.goto(0,-60)
    t.write(f'{m}',False,align="center", font=("Arial", 100, "bold"))
    time.sleep(2)
    t.bye()
def score_update(m): #점수표시
    score_pen.clear()
    score_pen.write(f"{m}  점수:{score} 시도:{attempt}",False,align="center", font=("Arial", 20, "bold"))

def play(x,y):
    global click_num,first_pick,second_pick,score,attempt

    if attempt==12:
        result('Game Over')
    else:
        click_num+=1 #클릭한 횟수를 체크(매 2회클릭마다 정답체크)
        #클릭한이미지를 먼저찾는다(함수를 만들어사용find_card)
        card_idx=find_card(x,y)
        turtles[card_idx].shape(img_list[card_idx])

        if click_num==1:
            first_pick=card_idx
        elif click_num==2:
            second_pick=card_idx
            click_num=0
            attempt+=1

            if img_list[first_pick]==img_list[second_pick]:
                score+=1
                score_update("정답")
                if score==8:
                    result('성공')
            else:
                #오답처리
                score_update("오답")
                turtles[first_pick].shape(default_img)
                turtles[second_pick].shape(default_img)

t.bgcolor("pink")
t.setup(700,700)
t.up()
t.ht()  #거북이를 숨긴다  ht()는 hide turtle/ st()는 show turtle
t.goto(0,280)
t.write("카드 짝 맞추기 게임",False,align="center", font=("Arial", 20, "bold"))

#점수 펜 객체 생성
score_pen=t.Turtle()
score_pen.up()  
score_pen.ht()
score_pen.goto(0,230)

#터들객체생성
turtles=[]
pos_x=[-210,-70,70,210]
pos_y=[-250,-110,30,170]

for x in range(4):
    for y in range(4):
        new_turtle=t.Turtle()
        new_turtle.up()
        new_turtle.color('pink')
        new_turtle.speed(0)
        new_turtle.goto(pos_x[x],pos_y[y])
        turtles.append(new_turtle)

#카드이미지
default_img='images_card/default_img.gif'
t.addshape(default_img)
img_list=[]
for i in range(8):
    img=f'images_card/img{i}.gif'
    t.addshape(img)
    img_list.append(img)  #같은 사진을 두장씩 넣는다.
    img_list.append(img)
    #카드이미지

random.shuffle(img_list)  #카드이미지를 섞는다.
for i in range(16):
    turtles[i].shape(img_list[i])

#1,2,3카운트를한다.
count=t.Turtle()
for i in range(3,0,-1):
    count.up()
    count.color('black')
    count.goto(0,0)
    count.write(f"{i}",False,align="center", font=("Arial", 150, "bold"))
    winsound.Beep(523,300)
    time.sleep(1)  #이후에 이미지를 숨긴다.
    count.clear()

for i in range(16): #모든 카드를 숨긴다.
    turtles[i].shape(default_img)

#이제부터 게임을 시작한다.
#첫번째 클릭했는지 두번째 클릭했는지 구분하는 변수를 만든다.
click_num=0  #클릭한 횟수를 체크(매 2회클릭마다 정답체크)
score=0
attempt=0 #시도횟수
first_pick=""  #첫번째 클릭한 카드
second_pick=""  #두번째 클릭한 카드

t.onscreenclick(play)  #마우스클릭이벤트를 만든다..



t.done()  # Close the turtle graphics window
```
