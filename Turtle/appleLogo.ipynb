```python
import turtle

def draw_apple_logo():
    screen = turtle.Screen()
    screen.bgcolor("white")
    screen.title("Apple Logo")

    pen = turtle.Turtle()
    pen.speed(3)
    pen.color("black")
    pen.hideturtle()

    # Draw the main body of the apple
    pen.penup()
    pen.goto(0, -100)
    pen.pendown()
    pen.begin_fill()
    pen.circle(100)
    pen.end_fill()

    # Draw the bite
    pen.penup()
    pen.goto(50, 50)
    pen.pendown()
    pen.color("white")
    pen.begin_fill()
    pen.circle(40)
    pen.end_fill()

    # Draw the leaf
    pen.penup()
    pen.goto(-30, 120)
    pen.setheading(45)
    pen.color("black")
    pen.pendown()
    pen.begin_fill()
    for _ in range(2):
        pen.circle(30, 90)
        pen.circle(30 // 2, 90)
    pen.end_fill()

    screen.mainloop()

draw_apple_logo()
```
