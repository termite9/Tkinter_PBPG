> 터들로 함수를 사용하여  매개변수를 사용하여 꽃잎갯수,사이즈,각도를 랜덤으로 불러와 꽃잎그리기

```python
#터들로 함수를 사용하여  매개변수를 사용하여 꽃잎갯수,사이즈,각도를 랜덤으로 불러와 꽃잎그리기
import turtle as t
import time
import random


t.speed(0) #1~10까지 속도조절가능하며  '0'은 최고속도
def petal(size,angle):
    for i in range(2):
        t.circle(size,angle)  #반지름을 넣는다(반지름,각)(크기 ㄱ,각도)
        t.left(180-angle)

def draw_flower(petal_num,petal_size,petal_angle): #매개변수를 사용
    t.color("pink")
    t.begin_fill()
    for i in range(petal_num):
        petal(petal_size,petal_angle)
        t.left(360/petal_num)
    t.end_fill()

    t.goto(0,-30)
    t.color("green")
    t.begin_fill()
    t.circle(30)
    t.end_fill()

t.bgcolor("black")

petal_num=random.randint(5,10)
petal_size=random.randint(50,150)
petal_angle=random.randint(70,120)
draw_flower(petal_num,petal_size,petal_angle)
t.ht()  # 거북이를 숨긴다
t.done()  # 
```

```python
#터들로 함수를 사용하여  매개변수를 사용하여 꽃잎갯수,사이즈,각도를 랜덤으로 불러와 꽃잎그리기에서 반환값을 사용하여 꽃잎그리기

import turtle as t
import time
import random

t.speed(0) #1~10까지 속도조절가능하며  '0'은 최고속도
t.bgcolor("black")
t.ht()  # 거북이를 숨긴다
t.colormode(255)  #색상모드를 255로 설정해줘야 RGB를 사용할수있다


def random_color():
    r=random.randint(0,255)
    g=random.randint(0,255)
    b=random.randint(0,255)
    return r,g,b

#꽃잎한장을 표현하는 함수
def petal():
    for i in range(2):
        t.circle(150,110)  #반지름을 넣는다(반지름,각)
        t.left(70)

def draw_flower():
    t.color(random_color()) #랜덤색상을 만든다 RGB를 사용하여 랜덤색상을 만든다.
    t.begin_fill()
    for i in range(10):
        petal()
        t.left(360/10)
    t.end_fill()

    t.goto(0,-30)
    t.color(random_color())
    t.begin_fill()
    t.circle(30)
    t.end_fill()

draw_flower()

t.done()  # 

```

> #터들로 간단하게 그림을 그리는 방법
#마우스를 드래그해서 그림을 그린다음 색상을 채운다.

```python
#터들로 간단하게 그림을 그리는 방법
#마우스를 드래그해서 그림을 그린다음 색상을 채운다.

import turtle as t
import time

def pen_up():
    t.penup()
def pen_down():
    t.pendown()

def begin_fill():
    t.begin_fill()
def end_fill():
    t.end_fill()

def green_color():
    t.color("green")
def pink_color():
    t.color("pink")
def black_color():
    t.color("black")
def white_color():
    t.color("white")

def draw_ondrag(x,y):
    t.ondrag(None)
    t.goto(x,y)
    t.ondrag(draw_ondrag)

t.speed(0) #1~10까지 속도조절가능하며  '0'은 최고속도
t.bgcolor("ivory")
t.pensize(3)

# t.onscreenclick(t.goto)  #마우스를 클릭할때마다 이벤트를 발생시킨다.
t.onscreenclick(draw_ondrag)  #마우스를 클릭할때마다 이벤트를 발생시킨다.
# t.onkeypress(함수,키)
t.onkeypress(green_color,"g")  #g키를 누르면 green_color함수를 실행한다.
t.onkeypress(pink_color,"p")  #p키를 누르면 pink_color함수를 실행한다.
t.onkeypress(black_color,"b")  #b키를 누르면 black_color함수를 실행한다.
t.onkeypress(white_color,"w")  #w키를 누르면 white_color함수를 실행한다.
t.onkeypress(pen_up,"Up")  #u키를 누르면 pen_up함수를 실행한다. Up은 방향키 위쪽을 의미한다.
t.onkeypress(pen_down,"Down")  #d키를 누르면 pen_down함수를 실행한다.Down은 방향키 아래쪽을 의미한다.

#색을 채우기 위해서는 그림을 그리기 전에 비긴필을 선택하고 그림을 그린후에 엔드필을 선택해야한다.
t.onkeypress(begin_fill,"Left")  #f키를 누르면 begin_fill함수를 실행한다. Lqft는 방향키 왼쪽을 의미한다.
t.onkeypress(end_fill,"Right")  #e키를 누르면 end_fill함수를 실행한다. Right는 방향키 오른쪽을 의미한다.
t.listen()  #키보드 이벤트를 받을수 있게 한다.
#위두개는 반드시 같이 사용해야한다. 

t.done()  # 
```

> 터들을 사용하여 간단한 그림판을 만든다. 지역변수와 전역변수를 사용한다.

```python
#터들을 사용하여 간단한 그림판을 만든다. 지역변수와 전역변수를 사용한다.

import turtle as t
import random
t.speed(0) #1~10까지 속도조절가능하며  '0'은 최고속도

t.up()
t.goto(0, 250)
t.write("그림판", False, align="center", font=("Arial", 25, "bold"))
t.down()

size=50
#space바를 누를때마다 랜덤색상이 나오도록 하기위해서 우선 리스트를 만들어준다,
colors=["red","green","blue","yellow","purple","orange","white","gray","brown","pink"]

def draw_dot(x,y): #매개변수가 1일 경우
    t.up()
    t.goto(x,y)
    t.dot(size)

def draw_triangle(x,y): #매개변수가 1일 경우
    t.up()
    t.goto(x,y)
    t.down()
    t.begin_fill()
    for i in range(3):
        t.forward(size)
        t.left(360/3)
    t.end_fill()

def draw_square(x,y): #매개변수가 1일 경우
    t.up()
    t.goto(x,y)
    t.down()
    t.setheading(random.randint(0,360))  #거북이가 바라보는 각도를 0으로 설정한다.
    t.begin_fill()
    for i in range(4):
        t.forward(size)
        t.left(360/4)
    t.end_fill()

def rand_color():
    t.color(random.choice(colors))

def size_up():
    global size #전역변수를 사용하기 위해서 global을 사용한다.
    size+=10

def size_down():
    global size #전역변수를 사용하기 위해서 global을 사용한다.
    size-=10
    if size <=10:
        size=10

# t.onscreenclick(함수,버튼)  #마우스를 클릭할때마다 이벤트를 발생시킨다.
#버튼이 1이면 왼쪽마우스, 2이면 가운데마우스, 3이면 오른쪽마우스가 선택된다.

t.bgcolor("ivory")
t.onscreenclick(draw_dot,1)  #마우스를 클릭할때마다 이벤트를 발생시킨다.
t.onscreenclick(draw_triangle,2)  #마우스를 클릭할때마다 이벤트를 발생시킨다.
t.onscreenclick(draw_square,3)  #마우스를 클릭할때마다 이벤트를 발생시킨다.
t.onkeypress(rand_color,"space")  #q키를 누르면 bye함수를 실행한다.
t.onkeypress(size_up,"Up")  #q키를 누르면 bye함수를 실행한다.
t.onkeypress(size_down,"Down")  #q키를 누르면 bye함수를 실행한다.
t.listen()  #키보드 이벤트를 받을수 있게 한다.
t.done()  #
```

```python
#터들을 이용하여 독초를 제거한다.터들을 이용하여 먹이를 먹는다.
import turtle as t
import random

def turn_left():
    player.left(30)
def turn_right():
    player.right(30)

def rand_pos():
    x_cor=random.randint(-300,300)
    y_cor=random.randint(-300,300)
    return x_cor,y_cor

#환경설정
t.setup(800,800) #화면크기
t.bgcolor("skyblue")
t.ht()  #거북이를 숨긴다  ht()는 hide turtle/ st()는 show turtle
t.up()  #붓을 종이에서 떼었다고 생각하면 된다. 그림이 그리지 않는다.

#변수
player_speed = 5
score = 0
game_over = False

#점수표시
t.goto(0,350)
t.write(f"Score: {score}", False, align="center", font=("Arial", 24, "bold") )


#플레이어
player = t.Turtle()
player.shape("turtle")
player.shapesize(2)
player.color("lavender")
player.up()
player.speed(0)

#먹이
food=t.Turtle()
food.ht()
food.shape("triangle")
food.color("darkgreen")
food.up()
food.speed(0)
food.setheading(90)
food.goto(rand_pos())
food.st()


#독초에 닿으면 게임이 종료되게한다.
p_hebs=t.Turtle()
p_hebs.ht()
p_hebs.shape("triangle")
p_hebs.color("red")
p_hebs.up()
p_hebs.speed(0)
p_hebs.setheading(90)
p_hebs.goto(rand_pos())
p_hebs.st()


#키보드를 누를경우 방향이 바뀌게 하는 로직
# t.onkeypress(함수,키)
t.onkeypress(turn_left,'Left')
t.onkeypress(turn_right,'Right')
t.listen() #키보드 이벤트를 받을수 있게 한다.

while not game_over:

    #플레이어 이동
    player.forward(player_speed)

    #플레이어 이동제한
    if player.xcor() > 360 or player.xcor() < -360 or player.ycor() > 360 or player.ycor() < -360: #벽면에 부딪히면
        player.right(180)

    #먹이에 닿았을때
    if player.distance(food) < 20: #거리가 20보다 작으면
        food.goto(rand_pos())
        p_hebs.goto(rand_pos())
        score += 1
        player_speed += 1
        t.clear()
        t.write(f"Score: {score}", False, align="center", font=("Arial", 24, "bold"))
    
    #독초가 닿을경우 게임오버가 되게한다.
    if player.distance(p_hebs) < 20: #거리가 20보다 작으면
        game_over = True
        t.goto(0,0)
        t.write("Game Over", False, align="center", font=("Arial", 24, "bold"))


t.done()
```
