```python
#밤하늘에 별그리기(다양하게 별의 색깔과 크기를 다르게 그리기)
import turtle as t
import time
import random


def tree(length,pen_size): #매개변수(가지길이, 가지굵기)

    angle=random.randint(15,30) #정수를 추출할때
    branch=random.uniform(0.6,0.9) #실수를 추출할때
    r_color=random.choice(color_list) #잎의색깔을 랜덤으로 선택한다.

    #가지 길이가 35보다 작을때 부터 잎을 그린다.    
    if length <35:
        t.color(r_color)
        t.stamp()  #잎을 찍는다.
        t.color("brown") #가지색깔을 갈색으로 설정한다.

    if length > 30:
        #가지그리기
        t.pensize(pen_size)
        pen_size *=0.7 #가지를 점점 가늘게 그리기위해서 가지굵기를 줄인다.
        t.forward(length)  
        #50도를 회전하여 왼쪽가지그린 후 오른쪽 가지를 그린다.
        t.left(angle)  #왼쪽으로 50도 회전
        tree(length*branch,pen_size)  #가지의 길이를 0.7배 줄인다.
        t.right(angle*2) #오른쪽으로 100도 회전
        tree(length*branch,pen_size)  #가지의 길이를 0.7배 줄인다.
        #원위치
        t.left(angle)  #왼쪽으로 50도 회전
        t.backward(length)  #100만큼 후진

def grass(): #바닥에 잔디를 그린다.
    t.up()
    t.goto(-380,-280)
    t.shapesize(7)
    while t.xcor() < 380:
        t.setheading(0)
        t.forward(10)
        t.color(random.choice(color_list))
        t.setheading(random.randint(60,120))
        t.stamp()

def star():
    for i in range(30):
        t.up()  #붓을 종이에서 떼었다고 생각하면 된다. 그림이 그리지 않는다.
        t.goto(random.randint(-300,300),random.randint(100,300))
        t.down()  #붓을 종이에 붙인다.
        t.color('white')
        t.begin_fill()
        star_size = random.randint(5,10)
        for i in range(5):
            t.forward(star_size)
            t.right(144)
        t.end_fill()

    t.ht()  #거북이를 숨긴다  ht()는 hide turtle/ st()는 show turtle
    # t.done()  # Close the turtle graphics window


color=("red","green","blue","yellow","purple","orange","white","gray","brown","pink")

t.bgcolor("black")

t.speed(0) #1~10까지 속도조절가능하며  '0'은 최고속도

star()

#잎을 그린다
color_list=["green",'darkgreen','lightgreen','yellowgreen','forestgreen','limegreen','seagreen','mediumseagreen','springgreen','palegreen','olive','olivedrab','darkolivegreen','greenyellow','chartreuse','lawngreen','mediumspringgreen']

t.speed(0) #1~10까지 속도조절가능하며  '0'은 최고속도
t.setheading(90)  #거북이가 바라보는 각도를 0으로 설정한다.

grass()
for i in range(20):
    
    t.up()
    t.goto(random.randint(-250,250),-300) #시작위치를 아래로 내린다.
    t.down()
    t.color('brown')
    tree(random.randint(50,100),random.randint(3,7))



t.done()
```
