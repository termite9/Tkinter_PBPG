> 밤하늘에 별그리기(다양하게 별의 색깔과 크기를 다르게 그리기)


```python
#밤하늘에 별그리기(다양하게 별의 색깔과 크기를 다르게 그리기)
import turtle as t
import time
import random

color=("red","green","blue","yellow","purple","orange","white","gray","brown","pink")

t.bgcolor("black")

t.speed(0) #1~10까지 속도조절가능하며  '0'은 최고속도
for i in range(100):
    t.up()  #붓을 종이에서 떼었다고 생각하면 된다. 그림이 그리지 않는다.
    t.goto(random.randint(-300,300),random.randint(-300,300))
    t.down()  #붓을 종이에 붙인다.
    t.color(random.choice(color))
    t.begin_fill()
    star_size = random.randint(10,50)
    for i in range(5):
        t.forward(star_size)
        t.right(144)
    t.end_fill()

    # t.color('black')
    # t.begin_fill()
    # for x in range(5):
    #     t.forward(star_size)
    #     t.right(144)
    # t.end_fill()
t.clear()  #화면을 지운다.
i=0
t.ht()  #거북이를 숨긴다  ht()는 hide turtle/ st()는 show turtle
    # t.done()  # Close the turtle graphics window
```
