> 터들을 이용하여 독초를 제거한다.터들을 이용하여 먹이를 먹는다.


```python
#터들을 이용하여 독초를 제거한다.터들을 이용하여 먹이를 먹는다.
import turtle as t
import random

def turn_left():
    player.left(30)
def turn_right():
    player.right(30)

def rand_pos():
    x_cor=random.randint(-300,300)
    y_cor=random.randint(-300,300)
    return x_cor,y_cor

#환경설정
t.setup(800,800) #화면크기
t.bgcolor("skyblue")
t.ht()  #거북이를 숨긴다  ht()는 hide turtle/ st()는 show turtle
t.up()  #붓을 종이에서 떼었다고 생각하면 된다. 그림이 그리지 않는다.

#변수
player_speed = 5
score = 0
game_over = False

#점수표시
t.goto(0,350)
t.write(f"Score: {score}", False, align="center", font=("Arial", 24, "bold") )


#플레이어
player = t.Turtle()
player.shape("turtle")
player.shapesize(2)
player.color("lavender")
player.up()
player.speed(0)

#먹이
food=t.Turtle()
food.ht()
food.shape("triangle")
food.color("darkgreen")
food.up()
food.speed(0)
food.setheading(90)
food.goto(rand_pos())
food.st()


#독초에 닿으면 게임이 종료되게한다.
p_hebs=t.Turtle()
p_hebs.ht()
p_hebs.shape("triangle")
p_hebs.color("red")
p_hebs.up()
p_hebs.speed(0)
p_hebs.setheading(90)
p_hebs.goto(rand_pos())
p_hebs.st()


#키보드를 누를경우 방향이 바뀌게 하는 로직
# t.onkeypress(함수,키)
t.onkeypress(turn_left,'Left')
t.onkeypress(turn_right,'Right')
t.listen() #키보드 이벤트를 받을수 있게 한다.

while not game_over:

    #플레이어 이동
    player.forward(player_speed)

    #플레이어 이동제한
    if player.xcor() > 360 or player.xcor() < -360 or player.ycor() > 360 or player.ycor() < -360: #벽면에 부딪히면
        player.right(180)

    #먹이에 닿았을때
    if player.distance(food) < 20: #거리가 20보다 작으면
        food.goto(rand_pos())
        p_hebs.goto(rand_pos())
        score += 1
        player_speed += 1
        t.clear()
        t.write(f"Score: {score}", False, align="center", font=("Arial", 24, "bold"))
    
    #독초가 닿을경우 게임오버가 되게한다.
    if player.distance(p_hebs) < 20: #거리가 20보다 작으면
        game_over = True
        t.goto(0,0)
        t.write("Game Over", False, align="center", font=("Arial", 24, "bold"))


t.done()
```
