> 터들을 이용하여 가위바위보게임을 만든다. 이미지는 gif파일로 사용한다


```python
#터들을 이용하여 가위바위보게임을 만든다. 이미지는 gif파일로 사용한다.
import turtle as t
import random
def com_choice(): #컴퓨터가 가위바위보를 선택하는 함수
    rand_choice=random.randint(0,2)
    com.shape=images[rand_choice]
    return com_list[rand_choice]

def result_print(user_c,com_c):
    global user_score,com_score
    t.clear()
    t.goto(0,-100)
    if user_c==com_c:
        t.write("무승부",False,align="center", font=("Arial", 50, "bold"))
    elif winning_case[user_c]==com_c:
        user_score+=1
        # user_pen.clear()
        # user_pen.write(user_score,False,align="center", font=("Arial", 20, "bold"))
        t.write("승",False,align="center", font=("Arial", 50, "bold"))
    else:
        com_score+=1
        # com_pen.clear()
        # com_pen.write(com_score,False,align="center", font=("Arial", 20, "bold"))
        
        t.write("패",False,align="center", font=("Arial", 50, "bold"))

def rock(x,y):
    user.shape(rock_gif)
    com=com_choice()
    result_print("rock",com)

def scissors(x,y):
    user.shape(scissors_gif)
    com=com_choice()
    result_print("scissors",com)

def paper(x,y):
    user.shape(paper_gif)
    com=com_choice()
    result_print("paper",com)

#기본설정
t.bgcolor("lavender")
t.title("가위바위보 게임")
t.ht()  #거북이를 숨긴다  ht()는 hide turtle/ st()는 show turtle
t.up()  #붓을 종이에서 떼었다고 생각하면 된다. 그림이 그리지 않는다.

rock_gif='images/rock.gif'
scissors_gif='images/scissors.gif'  #가위
paper_gif='images/paper.gif'  #보
t.addshape(rock_gif)
t.addshape(scissors_gif)        
t.addshape(paper_gif)

images=[rock_gif,scissors_gif,paper_gif]
com_list=['rock','scissors','paper']
winning_case={'rock':'scissors','scissors':'paper','paper':'rock'}
# user_score=0
# com_score=0

# #유저선택이미지
user=t.Turtle()
user.up()
user.speed(0)
user.goto(-200,200)
user.write("나의선택",False,align="center", font=("Arial", 20, "bold"))
user.goto(-200,-100)
user.shape(rock_gif)

#컴퓨터선택이미지
com=t.Turtle()
com.up()
com.speed(0)
com.goto(200,200)
com.write("컴퓨터선택",False,align="center", font=("Arial", 20, "bold"))
com.goto(200,-100)
com.shape(rock_gif)

# #유저점수펜
# user_pen=t.Turtle()
# user_pen.up()
# user_pen.ht()
# user_pen.goto(-200,100)
# user_pen.write(user_score,False,align="center", font=("Arial", 20, "bold"))
# com.goto(-200,100)
# com.shape(rock_gif)

# #컴점수펜
# com_pen=t.Turtle()
# com_pen.up()
# com_pen.ht()
# com_pen.goto(200,100)
# com_pen.write(user_score,False,align="center", font=("Arial", 20, "bold"))
# com.goto(200,100)
# com.shape(rock_gif)

# t.onscreenclick(함수,버튼)  #마우스를 클릭할때마다 이벤트를 발생시킨다.
#버튼이 1이면 왼쪽마우스, 2이면 가운데마우스, 3이면 오른쪽마우스가 선택된다.
t.onscreenclick(rock,1)  #좌클릭시 주먹을 낸다.
t.onscreenclick(scissors,2)  #가운데클릭시 가위를 낸다.
t.onscreenclick(paper,3)  #우클릭시 보를 낸다.




t.done()  # Close the turtle graphics window
```
