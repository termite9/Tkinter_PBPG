> 공 튕기기 애니메이션 만들기

```python
from tkinter import *
import random
import time


def move(): #공을 움직이는 함수 벽에 부딪치면 방향을 바꾼다
    global x_speed,y_speed
    ball_coor=canvas.coords(ball) #공의 좌표를 구한다.(x1,y1,x2,y2)
    #x1,y1은 공의 왼쪽 위 모서리의 좌표이고 x2,y2는 공의 오른쪽 아래 모서리의 좌표이다. 
    if ball_coor[3]>=HEIGHT or ball_coor[1]<=0:
        y_speed=-y_speed
    if ball_coor[2]>=WIDTH or ball_coor[0]<=0:
        x_speed=-x_speed
    canvas.move(ball,x_speed,y_speed)

window=Tk()
window.title("Bounce")

WIDTH=500
HEIGHT=300
canvas=Canvas(window,width=WIDTH,height=HEIGHT,bg='lightblue')
canvas.pack()

#공만들기
ball=canvas.create_oval(10,10,50,50,fill='pink')

#공에 움직임을 만든다.
x_speed=random.randint(5,20)
y_speed=random.randint(5,20)

while True:
    move()
    window.update()
    time.sleep(0.1)


window.mainloop()
```
