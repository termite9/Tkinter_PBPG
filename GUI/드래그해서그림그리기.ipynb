> 드래그해서 그림을 그리는 그림판 색깔선택및 굵기 선택 지우기 기능


> 전반부

```python
# 해당 프로그램의 전반부
from tkinter import *

def paint(event):
    x1,y1=event.x-1,event.y-1
    x2,y2=event.x+1,event.y+1
    canvas.create_oval(x1,y1,x2,y2,fill=pen_color,outline=pen_color)

def blue():
    global pen_color
    pen_color='blue'

def pink():
    global pen_color
    pen_color='pink'

pen_color='black'

window=Tk()
canvas=Canvas(window,width=500,height=500)
canvas.pack()

#그림을 그리는 기능(마우스를 드래그)
canvas.bind("<B1-Motion>",paint) #마우스왼쪽버튼을 누른채로 움직이면 paint함수가 호출됨

button_b=Button(window,text="파란색",width=5,bg="lightblue",command=blue)
button_b.place(x=0,y=0)
button_p=Button(window,text="핑크색",width=5,bg="pink",command=pink)
button_p.place(x=0,y=30)
window.mainloop()
```

> 중반부 

```python

from tkinter import *

def paint(event):
    x1,y1=event.x-1,event.y-1
    x2,y2=event.x+1,event.y+1
    canvas.create_oval(x1,y1,x2,y2,fill=pen_color,outline=pen_color,width=pen_size)

def blue():
    global pen_color
    pen_color='lightblue'

def pink():
    global pen_color
    pen_color='pink'

def change_color(color): #lambda함수를 사용
    global pen_color
    pen_color=color
def size_up():
    global pen_size
    pen_size+=1
def size_down():
    global pen_size
    if pen_size>1:
        pen_size-=1

pen_color='black'

window=Tk()
canvas=Canvas(window,width=500,height=500)
canvas.pack()

#그림을 그리는 기능(마우스를 드래그)
canvas.bind("<B1-Motion>",paint) #마우스왼쪽버튼을 누른채로 움직이면 paint함수가 호출됨

button_b=Button(window,text="파란색",width=5,bg="lightblue",command=blue)
button_b.place(x=0,y=0)
button_p=Button(window,text="핑크색",width=5,bg="pink",command=pink)
button_p.place(x=0,y=30)
button_r=Button(window,text="빨간색",width=5,bg="red",command=lambda:change_color("red"))
button_r.place(x=0,y=60)

#pen사이즈변경
pen_size=1
button_size_up=Button(window,text="굵게 +",command=size_up)
button_size_up.place(x=0,y=90)
button_size_down=Button(window,text="얇게 -",command=size_down)
button_size_down.place(x=0,y=120)

#지우기기능
button_e=Button(window,text="지우기",width=5,command=lambda:canvas.delete("all"))
button_e.place(x=0,y=150)

message=Label(window,text="마우스를 드래그하면 그림이 그려집니다.")
message.pack()
window.mainloop()
```
