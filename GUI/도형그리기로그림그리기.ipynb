> 캔버스 도형그리기 활용하여 그림그리기

```python
#캔버스 도형그리기 활용하여 그림그리기
from tkinter import *
window=Tk()

canvas=Canvas(window, width=500, height=500,bg="lightblue")
canvas.pack()

canvas.create_rectangle(20,20,480,480,fill="ivory",width=2)
canvas.create_text(250,100,text="!!Dream Big!!",font=("나눔바른펜",20,'bold'),fill="red")

# #귀 그리기
canvas.create_oval(200-10,190-50,200+30,190+50,fill="lightblue",width=2)

canvas.create_oval(300-10,190-50,300+30,190+50,fill="lightblue",width=2)

# #얼굴그리기
canvas.create_oval(250-100,250-80,250+100,250+80,fill="white",width=2)

# #코그리기
canvas.create_oval(250-10,300-8,250+10,300+8,fill="black",width=2)

#눈 만들기
canvas.create_arc(200-15,270-15,200+15,270+15,fill="black",width=2,start=60,style=ARC)
canvas.create_arc(300-15,270-15,300+15,270+15,fill="black",width=2,start=10,style=ARC)

canvas.create_line(20,330,480,330)

# #발그리기
canvas.create_oval(200-30,330-20,200+30,330+20,fill="pink",width=2)
canvas.create_oval(300-30,330-20,300+30,330+20,fill="pink",width=2)



window.mainloop()
```
