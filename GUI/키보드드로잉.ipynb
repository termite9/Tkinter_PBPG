> 키보드 드로잉

```python
from tkinter import *


def keydrawing(x,y):
    global cur_x, cur_y,line
    new_x=cur_x+x
    new_y=cur_y+y
    if draw or line==None:
        line=canvas.create_line(cur_x, cur_y, new_x, new_y, width=3)
    else:
        canvas.coords(line, cur_x, cur_y, new_x, new_y)
        #coords에 대한 설명은 다음 cell에서 확인할 수 있다.
    cur_x=new_x
    cur_y=new_y

def move_up(event):
    keydrawing(0,-5)
def move_down(event):
    keydrawing(0,5)
def move_left(event):
    keydrawing(-5,0)
def move_right(event):
    keydrawing(5,0)

def pen_up(event):
    global draw
    draw=False
def pen_down(event):
    global draw
    draw=True
#현재의 좌표값을 구한다.
cur_x=250
cur_y=250

draw=False #펜이 내려져 있는지 여부를 나타내는 변수로 False이면 펜이 들어져 있는 상태이다.
line=None #펜이 그리는 선을 나타내는 변수로 None이면 아직 선이 없다.
window = Tk()

window.title("키보드드로잉")
Label(window, text="u키로 펜을 들고 d키로 내립니다").pack()

canvas=Canvas(window, width=500, height=500)
canvas.pack()

#키보드화살표를 누르면 그림이 그려지도록 한다.
window.bind("<Up>", move_up)
window.bind("<Down>", move_down)
window.bind("<Left>", move_left)        
window.bind("<Right>", move_right)
window.bind("<u>", pen_up)
window.bind("<d>", pen_down)




window.mainloop()
```
