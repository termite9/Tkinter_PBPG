> 마우스 포인터에 닿으면 얼굴 표정이 변화한다.

```python
from tkinter import *

def change_to_img1(event):
    img_label.configure(image=img1)

def change_to_img2(event):
    img_label.configure(image=img2)
    
window=Tk()
window.title("마우스로 이미지 변경하기")
window.geometry("400x400")
window.configure(bg='pink')

img1=PhotoImage(file="images/face1.png")
img2=PhotoImage(file="images/face2.png")

#이미지1을 기본으로
img_label=Label(window, image=img1,bg='pink')
img_label.pack(pady=70)

#마우스에 닿으면 이미지2로 변경
img_label.bind("<Enter>", change_to_img1)
img_label.bind("<Leave>", change_to_img2)

window.mainloop()
```
