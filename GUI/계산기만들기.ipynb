> 계산기 만들기


```python
from tkinter import *


def clicked(digit):
    if digit=='⬅':
        #하나씩 숫자삭제
        input_entry.delete(len(input_entry.get())-1,END)
    else:
        input_entry.insert(END,digit)

def del_digit():
    input_entry.delete(0,END)
    result_label.config(text="")

def caculate():
    #입력된 수식을 계산하는 함수
    #잘못된수식일 경우 에러처리
    try:
        result=eval(input_entry.get()) #eval함수를 사용하면 입력값을 자동계산해준다
    except:
        result_label.config(text="수식이 잘못되었습니다")
    else:
        result_label.config(text=result)

IVORY = "#f8f8ff"
PINK= "#ffc0cb"
BLUE= "#add8e6"
GREEN= "#90ee90"


window=Tk()
window.title("계산기")
window.resizable(False, False)
window.config(padx=10,pady=10,bg=IVORY)

digits=[
    ['7','8','9','*'],
    ['4','5','6','/'],
    ['1','2','3','-'],
    ['C','0','⬅','+']
]

input_entry=Entry(window, width=30,font=("나눔바른펜", 20),bg=IVORY,justify='right')
input_entry.grid(row=0,column=0,columnspan=4)
input_entry.focus()


result_label=Label(window, text="", width=20,font=("나눔바른펜", 30),bg=GREEN)
result_label.grid(row=1,column=0,columnspan=4,pady=15)

for r in range(4):
    for c in range(4):
        digit=digits[r][c]
        button=Button(window, text=digit, width=8, font=("나눔바른펜", 15),bg=PINK,command=lambda x=digit:clicked(x))
        button.grid(row=r+2,column=c,  pady=2)

#입력창을 모두 지우는 버튼
clear_button=Button(window, text="c", width=17, font=("나눔바른펜", 15,'bold'),bg=BLUE,command=del_digit)
clear_button.grid(row=6,column=0,columnspan=2, pady=5)

#계산 버튼
cal_button=Button(window, text="=", width=17, font=("나눔바른펜", 15,'bold'),bg=BLUE,command=caculate)
cal_button.grid(row=6,column=2,columnspan=2, pady=5)

window.mainloop()
```
