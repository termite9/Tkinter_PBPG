> 케릭터 움직이기

```python
from tkinter import *

def player_move(event): #bind를 사용할때는 event를 넣어줘야함
    key=event.keysym
    if key=="Up":
        canvas.move(player,0,-10)
    elif key=="Down":
        canvas.move(player,0,10)
    elif key=="Left":
        canvas.move(player,-10,0)
    elif key=="Right":
        canvas.move(player,10,0)

window=Tk()
window.title("케릭터움직이기")

canvas=Canvas(window, height=500, width=500,bg='#FFEDD3')
canvas.pack()

img=PhotoImage(file="images/robot.png")
player=canvas.create_image(150,150,image=img,anchor=NW)

canvas.bind_all("<KeyPress>",player_move)
window.mainloop()
```
