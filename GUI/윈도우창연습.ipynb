> 윈도우 창 연습

```python
from tkinter import *

window=Tk()


#크기변경
window.geometry("300x200")
window.title("윈도우 창 연습")

label1=Label(window, text="안녕하세요", font=("나눔고딕", 10), fg="blue")

#배치방법 pack:자동으로 위에서 아래로 순서대로 하나씩 배치
label1.pack()

def button_click():
    label2.config(text="버튼이 눌렸습니다")
    label2.config(bg="yellow")

def change_color():
    label2.config(bg="white")
#버튼
button1=Button(window, text="클릭하세요", fg="red", command=button_click)   
button1.pack()

label2=Label(window, text="", font=("궁서체", 10), fg="blue")
label2.pack()


window.mainloop()
```
