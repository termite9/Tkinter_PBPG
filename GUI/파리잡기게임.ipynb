> 버튼을 클릭해서 날아다니는 파리를 잡는 게임

```python
from tkinter import *
from tkinter import messagebox
import random


next_move=None #다른 함수에 있는 변수를 사용하기 위해서 global로 설정한다

#버튼의 위치가 바뀌어지는 함수
def btn_move():
    global next_move
    new_x=random.randint(50, 450)
    new_y=random.randint(0,450)
    delay=random.randint(300,800) #300~600ms 사이의 랜덤한 시간을 설정
    button.place(x=new_x, y=new_y)

    next_move=window.after(delay, btn_move)  #200ms마다 btn_move함수를 호출

#버튼을 클릭했을 때 실행되는 함수
def clicked():
    window.after_cancel(next_move)  #버튼을 클릭하면 버튼의 이동을 멈춤
    messagebox.showinfo("알림", "버튼을 클릭했습니다.")
    window.destroy()

window=Tk()
window.title("버튼클릭게임")
window.geometry("500x500+100+100")
window.resizable(False, False)

img=PhotoImage(file='images/pari_img2.png')
button= Button(window, image=img,command=clicked)
button.place(x=250, y=250)

btn_move()

window.mainloop()
```
