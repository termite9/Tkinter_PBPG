> CM - INCH로 변환기

```python
from tkinter import *

def cm_to_inch():
    cm=float(Entry_input.get())
    inch=round(cm*0.3937,1)
    label_result.config(text=f"{cm}cm은 {inch}inch 입니다.")
    Entry_input.delete(0,END)

window=Tk()

window.title("cm/inch로변환")
window.geometry("250x300")
window.config(padx=30,pady=30,bg="lightblue")

lable_title=Label(window, text="cm/inch변환", font=("나눔고딕", 10,"bold"), fg="black",bg="white")
# lable_title.pack()
lable_title.grid(row=0,column=0,columnspan=3)

Entry_input=Entry(window, width=5)
# Entry_input.pack()
Entry_input.focus()
Entry_input.grid(row=1,column=0)

label_input=Label(window, text="cm", font=("나눔고딕", 10), fg="blue")
# label_input.pack()
label_input.grid(row=1,column=1)

button=Button(window, text="변환하기", fg="red" ,bg="black",\
    font=("나눔고딕", 10,"bold"), width=20 \
        ,command=cm_to_inch)
button.grid(row=2,column=0,columnspan=2)

label_result=Label(window, text="", font=("나눔고딕", 10), fg="blue")
label_result.grid(row=3,column=0,columnspan=2,pady=10)



window.mainloop()
```
