```python
#이미지 겔러리 만들기(이전과 다음버튼만 사용)
from tkinter import *
import random


def next_img():
    global img_num
    img_num+=1
    if img_num>7:
        img_num=0
    canvas.itemconfig(can_img,image=img_list[img_num])

def pre_img():
    global img_num
    img_num-=1
    if img_num<0:
        img_num=len(img_list)-1
    canvas.itemconfig(can_img,image=img_list[img_num])


window=Tk()

window.title("갤러리")
window.config(padx=10,pady=10,bg="lightblue")

button1 = Button(window, text="◀이전", font=("나눔바른펜",15,"bold"), bg="lightblue", fg="hotpink",command=pre_img)
button1.grid(row=0,column=0)

button2 = Button(window, text="다음▶", font=("나눔바른펜",15,"bold"), bg="lightblue", fg="hotpink",command=next_img)
button2.grid(row=0,column=1)

canvas=Canvas(window, width=680, height=680, bg="ivory")
canvas.grid(row=1,column=0,columnspan=2)

img_list=[]
for i in range(8):
    img=PhotoImage(file=f"images/img{i}.png")
    img_list.append(img)

img_num=random.randint(0,7)
can_img=canvas.create_image(340,340,image=img_list[img_num])


window.mainloop()
```

> 이미지 겔러리 만들기(이전과 다음버튼만 사용)-자동재생과 멈춤버튼 추가

```python
#이미지 겔러리 만들기(이전과 다음버튼만 사용)-자동재생과 멈춤버튼 추가
from tkinter import *
import random

auto=None
def next_img():
    global img_num
    img_num+=1
    if img_num>7:
        img_num=0
    canvas.itemconfig(can_img,image=img_list[img_num])

def pre_img():
    global img_num
    img_num-=1
    if img_num<0:
        img_num=len(img_list)-1
    canvas.itemconfig(can_img,image=img_list[img_num])

def auto_gallery():
    global auto
    next_img()
    auto=window.after(1000,auto_gallery) #1초마다 실행

def stop():
    window.after_cancel(auto)

window=Tk()

window.title("갤러리")
window.config(padx=10,pady=10,bg="lightblue")

button1 = Button(window, text="◀이전", font=("나눔바른펜",15,"bold"), bg="lightblue", fg="black",command=pre_img)
button1.grid(row=0,column=0,sticky="e")

button2 = Button(window, text="다음▶", font=("나눔바른펜",15,"bold"), bg="lightblue", fg="black",command=next_img)
button2.grid(row=0,column=1,sticky="w")

label=Label(window, text='랜덤박스', font=("나눔바른펜",20,"bold"), bg="lightblue", fg="hotpink")
label.grid(row=0,column=2)

button3 = Button(window, text="자동재생", font=("나눔바른펜",15,"bold"), bg="lightblue", fg="hotpink",command=auto_gallery)
button3.grid(row=0,column=3,sticky="e")

button4 = Button(window, text="Stop▶", font=("나눔바른펜",15,"bold"), bg="lightblue", fg="blue",command=stop)
button4.grid(row=0,column=4,sticky="w")

canvas=Canvas(window, width=680, height=680, bg="ivory")
canvas.grid(row=1,column=0,columnspan=5)

img_list=[]
for i in range(8):
    img=PhotoImage(file=f"images/img{i}.png")
    img_list.append(img)

img_num=0 
can_img=canvas.create_image(340,340,image=img_list[img_num])


window.mainloop()
```
