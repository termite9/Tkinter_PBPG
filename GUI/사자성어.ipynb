> 사자성어  저장 불러오기 찾기

```python
from tkinter import *
from tkinter import messagebox


def data_open():
    with open("idioms.txt","r",encoding="utf8") as file:
        data=file.read()
    result_label.config(text=data) #결과창 초기화

def data_save():
    idiom=idiom_entry.get()
    meaning=meaning_entry.get()

    if len(idiom)==0 or len(meaning)==0:
        messagebox.showerror("오류","사자성어와 뜻을 입력하세요")
        return
    with open("idioms.txt","a",encoding="utf8") as file:
        file.write(f"{idiom}:{meaning}\n")
    result_label.config(text=f"저장완료\n {idiom}:{meaning}")
    idiom_entry.delete(0,END) #입력창 초기화
    meaning_entry.delete(0,END) #입력창 초기화

def data_find():
    word=idiom_entry.get()
    with open("idioms.txt","r",encoding="utf8") as file:
        for data in file:
            if word in data:
                result_label.config(text=data)
                return
        result_label.config(text="데이터를 찾을 수 없습니다.")
        return
window=Tk()
window.title("")

img=PhotoImage(file="images/robot.png")
img_label=Label(window,image=img)
img_label.grid(row=0,column=0,columnspan=3)

title_label=Label(window,text="사자성어",font=("나눔바른펜",20,'bold'))
title_label.grid(row=1,column=0,columnspan=3,pady=10)

idiom_label=Label(window,text="사자성어",width=10,font=("나눔바른펜",15,'bold'))
idiom_label.grid(row=2,column=0)
idiom_entry=Entry(window,width=50,font=("나눔바른펜",15,'bold'))
idiom_entry.grid(row=2,column=1)

meaning_label=Label(window,text="뜻",width=10,font=("나눔바른펜",15,'bold'))
meaning_label.grid(row=3,column=0,pady=10)
meaning_entry=Entry(window,width=50,font=("나눔바른펜",15,'bold'))
meaning_entry.grid(row=3,column=1)

button_open=Button(window,text="불러오기",width=10,font=("나눔바른펜",12,'bold'),bg='pink',command=data_open)
button_open.grid(row=4,column=0,pady=20)

button_save=Button(window,text="저장",width=10,font=("나눔바른펜",12,'bold'),bg='blue',command=data_save)
button_save.grid(row=4,column=1,pady=20)

button_find=Button(window,text="찾기",width=10,font=("나눔바른펜",12,'bold'),bg='pink',command=data_find)
button_find.grid(row=4,column=2,pady=20)


result_label=Label(window,wraplength=500,font=("나눔바른펜",12))#글자가 길어지면 자동으로 줄바꿈
result_label.grid(row=5,column=0,columnspan=3)

window.mainloop()
```
