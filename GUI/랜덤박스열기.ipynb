```python
from tkinter import *

import random

def rand_choice():
    r_img=random.choice(img_list)
    canvas.itemconfig(canvas_img,image=r_img)

window=Tk()
window.title("random Box")
window.config(padx=10,pady=10, bg="lightblue")

canvas=Canvas(window, width=680, height=680, bg="ivory")
canvas.pack()

img_main=PhotoImage(file="images/gift.png")
canvas_img = canvas.create_image(340,340,image=img_main)
canvas.create_text(340,30,text="랜덤박스",font=("나눔바른펜",30,"bold"),fill="brown")

img_list=[]
for i in range(8):
    img=PhotoImage(file=f"images/img{i}.png")
    img_list.append(img)

button = Button(window, text="랜덤박스열기", font=("나눔바른펜",20,"bold"), bg="white", fg="hotpink", command=rand_choice)
button.place(x=500,y=10)



window.mainloop()
```
