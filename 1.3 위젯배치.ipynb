#위젯을 배치할때 사용되는 메서드는 pack(), grid(), place()가 있다.
#pack()은 위젯을 블록처럼 배치하는 것이고, grid()는 격자에 배치하는 것이다. place()는 절대위치에 배치하는 것이다.
#grid()는 행과 열을 지정하여 위젯을 배치한다. 행과 열은 0부터 시작한다.

```python
# pack(옵션)
# pack() 메서드는 위젯을 화면에 배치하는 메서드이다. pack() 메서드는 다음과 같은 옵션을 가질 수 있다.
# fill : 위젯이 할당된 공간을 채우는 방법을 지정한다. 기본값은 NONE이다.
# expand : 위젯이 할당된 공간을 확장할지 여부를 지정한다. 기본값은 NO이다.
# side : 위젯을 화면에 배치할 방향을 지정한다. 기본값은 TOP이다.
# anchor : 위젯이 할당된 공간 내에서 위치를 지정한다. 기본값은 CENTER이다.
# padx : 위젯의 좌우 여백을 지정한다. 기본값은 0이다.
# pady : 위젯의 상하 여백을 지정한다. 기본값은 0이다.
# ipadx : 위젯 내부의 좌우 여백을 지정한다. 기본값은 0이다.
# ipady : 위젯 내부의 상하 여백을 지정한다. 기본값은 0이다.

import tkinter

window=tkinter.Tk()
window.geometry("640x400+100+100")
window.resizable(False, False) # 크기 조절 불가능

bi=tkinter.Button(window, text="top")
bi.pack(side='top')
b2=tkinter.Button(window, text="bottom")
b2.pack(side='bottom')
b3=tkinter.Button(window, text="left")
b3.pack(side='left')
b4=tkinter.Button(window, text="right")
b4.pack(side='right', anchor='n')
b5=tkinter.Button(window, text="center", bg="yellow")
b5.pack()


window.mainloop()
```

```python
#객체지향 스타일로 작성해보자.

import tkinter as tk

class Application(tk.Tk):
    def __init__(self):
        super().__init__()
        self.geometry("640x400+100+100")
        self.resizable(False, False)
        self.create_widgets()

    def create_widgets(self):

        self.b1=tk.Button(self, text="top")
        self.b1.pack(side='top')
        self.b2=tk.Button(self, text="bottom")
        self.b2.pack(side='bottom')
        self.b3=tk.Button(self, text="left")
        self.b3.pack(side='left')
        self.b4=tk.Button(self, text="right")
        self.b4.pack(side='right', anchor='n')
        self.b5=tk.Button(self, text="center", bg="yellow")
        self.b5.pack()

app=Application()
app.mainloop()
```

```python
# fill옵션의 의미를 파악해보자

import tkinter

window=tkinter.Tk()
window.geometry("640x400+100+100")
window.resizable(False, False) # 크기 조절 불가능

b1=tkinter.Button(window, text="top")
b2=tkinter.Button(window, text="bottom")
b3=tkinter.Button(window, text="left")
b4=tkinter.Button(window, text="right")
b5=tkinter.Button(window, text="center", bg="yellow")
b1.pack(side='top')
b2.pack(side='bottom',fill='x')
b3.pack(side='left',fill='y')
b4.pack(side='right', anchor='s')
b5.pack(fill='both')

window.mainloop()
```

```python
#할당되지 않은 범위를 채우는 방법을 알아보자  expand옵션을 사용하면된다.

import tkinter

window=tkinter.Tk()
window.geometry("640x400+100+100")
window.resizable(False, False) # 크기 조절 불가능

b1=tkinter.Button(window, text="top")
b2=tkinter.Button(window, text="bottom")
b3=tkinter.Button(window, text="left")
b4=tkinter.Button(window, text="right")
b5=tkinter.Button(window, text="expand로  확장함", bg="yellow")
b1.pack(side='top')
b2.pack(side='bottom',fill='x')
b3.pack(side='left',fill='y')
b4.pack(side='right', anchor='s')

b5.pack(fill='both', expand=True)

window.mainloop()
```

```python
# grid(옵션)
# grid() 메서드는 위젯을 행과 열로 배치하는 메서드이다. grid() 메서드는 다음과 같은 옵션을 가질 수 있다.
# row : 위젯의 행 위치를 지정한다. 기본값은 0이다.
# column : 위젯의 열 위치를 지정한다. 기본값은 0이다.
# rowspan : 위젯이 차지할 행의 개수를 지정한다. 기본값은 1이다.
# columnspan : 위젯이 차지할 열의 개수를 지정한다. 기본값은 1이다.
# sticky : 위젯이 셀 내에서 위치를 지정한다. 기본값은 CENTER이다. N, S, E, W, NE, NW, SE, SW 등의 값이 올 수 있다.

import tkinter

window=tkinter.Tk()
window.geometry("640x400+100+100")
# window.resizable(False, False) # 크기 조절 불가능

b1=tkinter.Button(window, text="(0,0)")
b2=tkinter.Button(window, text="(0,1)")
b3=tkinter.Button(window, text="(0,2)")
b4=tkinter.Button(window, text="(1,0)") 
b5=tkinter.Button(window, text="(1,1)")
b6=tkinter.Button(window, text="(2,2)")
b7=tkinter.Button(window, text="(2,3)")

b1.grid(row=0, column=0)
b2.grid(row=0, column=1)
b3.grid(row=0, column=2)
b4.grid(row=1, column=0,rowspan=2,sticky='s')
b5.grid(row=1, column=1, columnspan=2, sticky='e')
b6.grid(row=2, column=1)
b7.grid(row=2, column=2)


window.mainloop()
```

```python
# place(옵션)
# place() 메서드는 위젯을 절대 위치에 배치하는 메서드이다. place() 메서드는 다음과 같은 옵션을 가질 수 있다.
# x : 위젯의 x 좌표를 지정한다. 기본값은 0이다.
# y : 위젯의 y 좌표를 지정한다. 기본값은 0이다.
# width : 위젯의 너비를 지정한다.
# height : 위젯의 높이를 지정한다.
# anchor : 위젯이 할당된 공간 내에서 위치를 지정한다. 기본값은 NW이다.
# relx : 부모 위젯에 대한 상대적인 x 좌표를 지정한다. 0.0은 왼쪽 끝, 1.0은 오른쪽 끝을 의미한다.
# rely : 부모 위젯에 대한 상대적인 y 좌표를 지정한다. 0.0은 위쪽 끝, 1.0은 아래쪽 끝을 의미한다.
# relwidth : 부모 위젯에 대한 상대적인 너비를 지정한다.
# relheight : 부모 위젯에 대한 상대적인 높이를 지정한다.

from tkinter import *

top=Tk()
top.geometry("250x200+10+10")

L1=Label(top, text="Physics")
L1.place(x=10, y=10)
E1=Entry(top, bd=3)
E1.place(x=80, y=10)
L2=Label(top, text="Maths")
L2.place(x=10, y=40)
E2=Entry(top, bd=3)
E2.place(x=80, y=40)
L3=Label(top, text="Total")
L3.place(x=10, y=150)
E3=Entry(top, bd=3)
E3.place(x=80, y=150)

B=Button(top, text="Submit", bg="yellow", fg="red")
B.place(x=100, y=100)


top.mainloop()
```

```python
# 위젯의 크기를 상대적으로 지정해보자

from tkinter import *
top=Tk()
top.geometry("250x200+10+10")   
L1=Label(top, text="Physics")
L1.place(relx=0.05,rely=0.05,relwidth=0.2,relheight=0.15)
E1=Entry(top, bd=3)
E1.place(relx=0.3,rely=0.05,relwidth=0.65,relheight=0.15)
L2=Label(top, text="Maths")
L2.place(relx=0.05,rely=0.25,relwidth=0.2,relheight=0.15)
E2=Entry(top, bd=3)
E2.place(relx=0.3,rely=0.25,relwidth=0.65,relheight=0.15)
L3=Label(top, text="Total")
L3.place(relx=0.05,rely=0.75,relwidth=0.2,relheight=0.15)
E3=Entry(top, bd=3)
E3.place(relx=0.3,rely=0.75,relwidth=0.65,relheight=0.15)
B=Button(top, text="Submit", bg="yellow", fg="red")
B.place(relx=0.4,rely=0.5,relwidth=0.2,relheight=0.15)




top.mainloop()
```
