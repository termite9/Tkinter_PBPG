#### 2.8. SpinBox 위젯

- 증감버튼을 이용하여 원하는 값을 입력할 수 있는 위젯

- w=SpinBox(master,option=valuse)

> 옵션

- from_
- to
- values

```python
# 사용자가 SpinBox를 누를 때 마다 현재값을 밑의 레이블에 출력하는 프로그램

import tkinter as tk

window=tk.Tk()
window.title("SpinBox Example")
window.geometry("300x100")
window.resizable(False, False)

current_value=tk.StringVar(value=0)
spin_box=tk.Spinbox(window, from_=0, to=10, increment=1, textvariable=current_value)
spin_box.pack()
label=tk.Label(window, textvariable=current_value)
label.pack()

window.mainloop()
```
