#### 2.10. RadioButton 위젯

- 사용자에게 옵션을 받아들일때 사용한다.
- 외관은 원형이다
- 체크버튼과 다른 가장 큰 것은  라디오 버튼은 여러 항목 중 하나만 선택할 수 있다.

- w=RadioButton(master,option=value)

> 옵션

- command
- state
- textvariable
- value
- variable

> 메서드

- deselect()
- flash()
- invoke()
- select()


```python
import tkinter as tk

window=tk.Tk()
window.title("Check Button")
window.geometry("300x100")
window.resizable(width=False, height=False)

radiovar=tk.IntVar()
radiovar.set(1)
r1=tk.Radiobutton(window, text="Option 1", variable=radiovar, value=1)
r2=tk.Radiobutton(window, text="Option 2", variable=radiovar, value=2)  # value is the value of the variable when this radio button is selected
r3=tk.Radiobutton(window, text="Option 3", variable=radiovar, value=3)
r1.pack()
r2.pack()   
r3.pack()

window.mainloop()
```

> 라디오버튼 선택시 실행할 작업은 command옵션을 이용한면 된다 

```python
import tkinter as tk

def display_radiobuttons():
    val=radiovar.get()
    if val==1:
        print("Option 1 is selected")
    elif val==2:
        print("Option 2 is selected")
    elif val==3:
        print("Option 3 is selected")

window=tk.Tk()
window.title("Check Button")
window.geometry("300x100")
window.resizable(width=False, height=False)

radiovar=tk.IntVar()
radiovar.set(1)
r1=tk.Radiobutton(window, text="Option 1", variable=radiovar, value=1, command=display_radiobuttons)
r2=tk.Radiobutton(window, text="Option 2", variable=radiovar, value=2,command=display_radiobuttons)  
r3=tk.Radiobutton(window, text="Option 3", variable=radiovar, value=3,command=display_radiobuttons)
r1.pack()
r2.pack()   
r3.pack()

window.mainloop()
```

> 반복문을 사용하여 심플하게

```python
import tkinter as tk

def display_radiobuttons():
    val=radiovar.get()
    if val==1:
        print("한식")
    elif val==2:
        print("중식")
    elif val==3:
        print("일식")
    elif val==4:
        print("양식")

window=tk.Tk()
window.title("Check Button")
window.geometry("300x100")
window.resizable(width=False, height=False)

foods=[('한식',1),('중식',2),('일식',3),('양식',4)]

radiovar=tk.IntVar()
radiovar.set(1)

for food, val in foods:
    tk.Radiobutton(window, text=food, variable=radiovar, value=val, command=display_radiobuttons).pack()

window.mainloop()
```

> 음식옵션과 후식옵션에서 가각 하나씩 선택한 후에 버튼을 누르면 어떤 옵션이 선택되었는지 확인한다

```python
import tkinter as tk

def display_radiobuttons():
    val=radiovar1.get()
    if val==1:         print("한식")
    elif val==2:        print("중식")
    elif val==3:        print("일식")
    elif val==4:        print("양식")

    val=radiovar2.get()
    if val==1:
        print("coffee")
    elif val==2:
        print("녹차")
    elif val==3:
        print("iceCream")


window=tk.Tk()
window.title("Check Button")
window.geometry("220x300")
window.resizable(width=False, height=False)

foods=[('한식',1),('중식',2),('일식',3),('양식',4)]
desserts=[('coffee',1),('녹차',2),('iceCream',3)]

radiovar1=tk.IntVar()
radiovar1.set(1)
radiovar2=tk.IntVar()
radiovar2.set(1)
frm1=tk.LabelFrame(window, text="음식")
frm2=tk.LabelFrame(window, text="디저트")

for food, val in foods:
    tk.Radiobutton(frm1, text=food, variable=radiovar1, value=val).pack()
for dessert, val in desserts:
    tk.Radiobutton(frm2, text=dessert, variable=radiovar2, value=val).pack(anchor=tk.W)

frm1.pack(anchor=tk.CENTER)
frm2.pack(anchor=tk.CENTER)    

tk.Button(window, text="Order", command=display_radiobuttons).pack(anchor=tk.CENTER)


window.mainloop()
```
