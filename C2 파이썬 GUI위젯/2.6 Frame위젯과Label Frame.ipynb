#### 2.6 Frame 위젯과 Label Frame 위젯
- 관련된 위젯들을 시각적으로 하나의  그룹으로 표시할 때 사용한다

- w=Frame(master,option=value)
- w=LabelFrame(master,option=value)

> 추가옵션

- labelAnchor
- text

```python
#레이블 프레임 위젯과  프레임 위젯을 생성하고, 그 안에 세 개의 버튼을 배치하는 예이다
import tkinter as tk

window=tk.Tk()
window.title("Label Frame Example")
window.geometry("220x200")
window.resizable(False,False)

#label frame 생성
lf=tk.LabelFrame(window,text="Label Frame")
lf.grid(row=0,column=0,padx=20,pady=20)

#버튼 생성
left_button1=tk.Button(lf,text="Left1")
left_button1.grid(row=0,column=0,padx=10,pady=10)
center_button1=tk.Button(lf,text="Center1")
center_button1.grid(row=0,column=1,padx=10,pady=10)
right_button1=tk.Button(lf,text="Right1")
right_button1.grid(row=0,column=2,padx=10,pady=10)

#frame 생성
lf=tk.Frame(window)
lf.grid(row=1,column=0,padx=20,pady=20)

#버튼 생성
left_button2=tk.Button(lf,text="Left2")
left_button2.grid(row=0,column=0,padx=10,pady=10)
center_button2=tk.Button(lf,text="Center2")
center_button2.grid(row=0,column=1,padx=10,pady=10)
right_button2=tk.Button(lf,text="Right2")
right_button2.grid(row=0,column=2,padx=10,pady=10)


window.mainloop()
```
