#### 2.12. Scrolled위젯

- 스크롤링을 제공하는 위젯

- w=Scrollbar(master,option=value)


> 옵션

- jump
- orient

> 메서드

- get()
- set(first,last)

> 스트롤바를 부착할 수 있는 위젯

- Text, Entry, Canvas, Listbox 등이 있다.

   - 이들 위젯에는 공통적으로 xscrollcommand와 ycrollcommand 옵션이 있다.

```python
# listbox위젯에 스크롤바를 추가하는 방법
import tkinter as tk

window=tk.Tk()
window.title("Listbox with Scrollbar")
window.geometry("300x200")
window.resizable(False, False)

var1=tk.StringVar()
var1.set(('aaaaaaaaaaaaaaaaaaaaaaa','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p'))

yscroll=tk.Scrollbar(orient=tk.VERTICAL)
yscroll.grid(row=0, column=1, sticky=tk.N+tk.S)

xscroll=tk.Scrollbar(orient=tk.HORIZONTAL)
xscroll.grid(row=1, column=0, sticky=tk.E+tk.W)

lb=tk.Listbox(window, listvariable=var1, selectmode=tk.EXTENDED,yscrollcommand=yscroll.set, xscrollcommand=xscroll.set)
lb.grid(row=0, column=0, sticky=tk.N+tk.S+tk.W+tk.E)

xscroll['command']=lb.xview
yscroll['command']=lb.yview
window.mainloop()
```
