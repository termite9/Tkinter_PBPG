#### 2.7. Scale 위젯

- 주어진 값의 범위내에서 값을 입력받을때 사용하는 위젯으로 값을 선택할 수 있는 그래픽 슬라이더를 제공한다

- w=Scale(master,option=value)

> 추가옵션

- from_
- orient
- to

> 메서드

- get()
- set(value)

```python
#버튼을 누르면 Scale위젯의 현재값을 그 아래에 있는 레이블에 출력하는 프로그램이다.

import tkinter as tk

def sel():
    selection="값 = "+str(scale.get())
    label.config(text=selection)

window=tk.Tk()
window.title("Scale and Label")
window.geometry("220x200")
window.resizable(width=False, height=False)

scale=tk.Scale(window)
scale.pack(anchor="center")

button=tk.Button(window, text="현재값 ?", command=sel)
button.pack(anchor="center")

label=tk.Label(window)
label.pack(anchor="center")

window.mainloop()
```

- 또다른 방법으로 제어변수를 이용한다



```python

import tkinter as tk

def sel():
    selection="값 = "+str(scale.get())
    label.config(text=selection)

def init():
    var.set(0)
    label.config(text="값 = 0")

window=tk.Tk()
window.title("Scale Demo")
window.geometry("220x200")
window.resizable(width=False, height=False)

var = tk.DoubleVar()
scale=tk.Scale(window, variable=var)
scale.pack(anchor="center")

button=tk.Button(window, text="현재값 ?", command=sel)
button.pack(anchor="center")
button2=tk.Button(window, text="초기화", command=init)
button2.pack(anchor="center")

label=tk.Label(window)
label.pack(anchor="center")

window.mainloop()
```
