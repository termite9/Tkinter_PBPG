#### 2.9.  CheckButton 위젯

- 사용자로 부터 옵션을 입력 받는데 사용한다.

- w=CheckBox(master,option=value)

> 옵션

- command
- offvalue
- onvalue
- state
- variable

> 메서드

- deselect()
- flash()
- invoke()
- select()
- toggle()

```python
# 단순히 두개의 체크박스를 출력하는 프로그램
import tkinter as tk

window=tk.Tk()
window.title("Check Button")
window.geometry("300x100")
window.resizable(width=False, height=False)

checkVar1=tk.IntVar()
checkVar2=tk.IntVar()
C1=tk.Checkbutton(window, text="한식", variable=checkVar1, onvalue=1, offvalue=0, height=5, width=20)
C2=tk.Checkbutton(window, text="중식", variable=checkVar2, onvalue=1, offvalue=0, height=5, width=20)
C1.pack()
C2.pack()

window.mainloop()
```

```python
#위 예를 수정  체크박스 선택이나 해제 시에 현재의 상태를 출력해본다.

import tkinter as tk

def display_checkbutton1():
    if checkVar1.get()==1:
        print("한식 체크박스가 선택되었습니다.")
    else:
        print("한식 체크박스가 해제되었습니다.")

def display_checkbutton2():
    if checkVar2.get()==1:
        print("중식 체크박스가 선택되었습니다.")
    else:
        print("중식 체크박스가 해제되었습니다.")

window=tk.Tk()
window.title("Check Button")
window.geometry("300x100")
window.resizable(width=False, height=False)

checkVar1=tk.IntVar()
checkVar2=tk.IntVar()
C1=tk.Checkbutton(window, text="한식", variable=checkVar1, onvalue=1, offvalue=0, height=5, width=20, command=display_checkbutton1)
C2=tk.Checkbutton(window, text="중식", variable=checkVar2, onvalue=1, offvalue=0, height=5, width=20, command=display_checkbutton2)
C1.pack()
C2.pack()

window.mainloop()
```

```python
# 일반적으로는 체크박스를 통해 옵션을 선택한 후에 버튼을 누르면 특정 작업이 수행되도록 한다.

import tkinter as tk


def display_checkbuttons():
    if checkVar1.get()==1:
        print("한식 체크박스가 선택되었습니다.")
    else:
        print("한식 체크박스가 해제되었습니다.")
    if checkVar2.get()==1:
        print("중식 체크박스가 선택되었습니다.")
    else:
        print("중식 체크박스가 해제되었습니다.")

window=tk.Tk()
window.title("Check Button")
window.geometry("220x200")
window.resizable(width=False, height=False)

checkVar1=tk.IntVar()
checkVar2=tk.IntVar()
C1=tk.Checkbutton(window, text="한식", variable=checkVar1, onvalue=1, offvalue=0, height=5, width=20)
C2=tk.Checkbutton(window, text="중식", variable=checkVar2, onvalue=1, offvalue=0, height=5, width=20)
btn=tk.Button(window, text="click here", command=display_checkbuttons)
C1.pack()
C2.pack()
btn.pack()

window.mainloop()
```
