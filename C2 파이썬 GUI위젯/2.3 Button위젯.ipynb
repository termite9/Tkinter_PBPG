- 마우스 클릭으로 특정 업무를 수행시킬 수 있게 한다.
- w=Button(master,option=value)
- command옵션은 마우스 왼쪽 버튼을 누를 경우 자동으로 실행되는 콜백함수를 지정한다.
- bind()와 다른 점은 해당 좌표를 구할 수 없다는 것이다.

> 메서드
- flash()
- invoke()

```python
# + 버튼을 누르면 숫자가 1씩 증가하고, - 버튼을 누르면 숫자가 1씩 감소하고 *버튼을 누르면 숫자가 무작위로 증가되거나 감소한다

import random
import tkinter as tk

def inc():
    c=cnt_label.cget('text')
    c+=1
    cnt_label.config(text=c)

def dec():
    c=cnt_label.cget('text')    
    c-=1
    cnt_label.config(text=c)
def rand():
    b=random.choice([True,False])
    if (b):
        inc_btn.invoke()
    else:
        dec_btn.invoke()

window=tk.Tk()
window.title('my window')

cnt_label=tk.Label(window,text=0)
cnt_label.grid(row=0,column=0,columnspan=3)
inc_btn=tk.Button(window,text='+',command=inc)
inc_btn.grid(row=1,column=0)
dec_btn=tk.Button(window,text='-',command=dec)
dec_btn.grid(row=1,column=1)
rand_btn=tk.Button(window,text='*',command=rand)
rand_btn.grid(row=1,column=2)


window.mainloop()
```
