#### Entry 위젯
- 텍스트를 표시하고 입력할 수 있게 해주는 위젯으로 사용자로 부터 텍스트 입력을 받을 경우사용
- w=Entry(master,option=value)

> option
- show
- state


```python
#로그인 화면을 작성해본다
import tkinter as tk


def submit():
    name=name_entry.get()
    password=password_entry.get()
    print("name : "+name)
    print("password : "+password)
    name_entry.config(text="")
    password_entry.config(text="")

window=tk.Tk()
window.geometry("220x80")


name_label=tk.Label(window,text="로그인")
name_label.grid(row=0,column=0)
name_entry=tk.Entry(window)
name_entry.grid(row=0,column=1)

password_label=tk.Label(window,text="비밀번호")
password_label.grid(row=1,column=0)
password_entry=tk.Entry(window,show="*")
password_entry.grid(row=1,column=1)

sub_button=tk.Button(window,text="Submit",command=submit)
sub_button.grid(row=2,column=0,columnspan=2)


window.mainloop()
```

> 메서드
- delete(first,lat=None)
- get()
- icursor(index)
- insert(index,s)
- select_clear()
- select_range(start,end)



```python
import tkinter as tk

def set_text(text):
    e.delete(0,tk.END)
    e.insert(0,text)
    return

window=tk.Tk()

e=tk.Entry(window,width=20)
e.pack()

b1=tk.Button(window,text="석류",command=lambda:set_text("석류"))
b1.pack()

b2=tk.Button(window,text="사과",command=lambda:set_text("사과"))
b2.pack()

window.mainloop()
```

> Entry는 Tkinter제어변수에 의해서도 값을 읽거나 지정할 수있다.
> 제어변수는 저장하려는 값 유형에 따라 다음 클래스 생성자 중 하나를 사용하여 생성하면 된다.
- v=tk.DoubleVar()   #float값 저장. 기본값은 0.0
- v=tk.IntVar()
- v=tk.StringVar()   #문자열 값 저장. 기본값은 ''

```python
import tkinter as tk

window=tk.Tk()

entry_text=tk.StringVar()
e=tk.Entry(window,width=10,textvariable=entry_text)
e.pack()

b1=tk.Button(window,text="석류",command=lambda:entry_text.set("석류"))
b1.pack()

b2=tk.Button(window,text="사과",command=lambda:entry_text.set("사과"))
b2.pack()
window.mainloop()
```
