#### Text위젯
- 여러줄 입력을 가능하게 한다
- 칼라와 폰트도 변경이 가능하다
- 일부 텍스트만 변경도 가능하다
- 심지어 윈도우이미지도 임베딩이 가능하다.
- w=Text(master,option=value)

> option

> 메서드

```python
import tkinter as tk

window = tk.Tk()
window.title("Hello World")

T=tk.Text(window, height=2, width=30)
T.pack()
T.insert(tk.END, "국민의당과 합당이 사실상\n 물건너갈 것이라는 뜻으로 읽힌다.\n Hello World")

window.mainloop()
```

- 좁은공간 내에서 긴 메세지를 출력하기 위해 스크롤바를 부착한다.

```python
import tkinter as tk

window = tk.Tk()
window.title("Hello World")

VS=tk.Scrollbar(window)
VS.pack(side=tk.RIGHT, fill=tk.Y)

T=tk.Text(window, height=2, width=30)
VS.config(command=T.yview)
T.config(yscrollcommand=VS.set)
T.pack()

T.insert(tk.END, "국민의당과 합당이 사실상\n 물건너갈 것이라는 뜻으로 읽힌다.\n Hello World")
window.mainloop()
```
