#### 2.11. Listbox 위젯

- 목록상자는 항목들을 여러개 보여주고 이 중에서 여러개 선택할 수 있도록 해주는 위젯이다.

- w=Listbox(master,option=value)

> 옵션

- hightlightcolor
- listvariable
- selectmode
  - browse
  - single
  - multiple
  - extended

> 메서드

- active(index)
- curselection()
- delete(first,last=None)
- get(first,last=None)
- insert(index,*elements)
- see(index)
- size()

> 다음은 리스트박스에 항목을 추가하거나 제거하는 방법과 항목 선택 파악방법을 보여준다

```python
import tkinter as tk

def print_selection():
    value= lb.get(lb.curselection())
    print(value)

window = tk.Tk()
window.title('my window')
window.geometry('220x220')
window.resizable(width=False, height=False)

b1=tk.Button(window, text='click me', width=15, height=2, command=print_selection)
b1.pack()

var1=tk.StringVar()
var1.set(('a', 'b', 'c', 'd'))
lb=tk.Listbox(window, listvariable=var1)
list_items=['1', '2', '3', '4']
for item in list_items:
    lb.insert('end', item)
lb.insert(1, 'first')
lb.insert(2, 'second')      
lb.delete(2)
lb.pack()

window.mainloop()
```
